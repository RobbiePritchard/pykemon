```python
filename = 'moves.txt'
with open(filename, 'r') as f:
    for line in f:
        line=line.strip()
        line = line.split()
        if len(line) == 10:
            a = line[1]
            b = line[2]
            line[2] = a +' ' + b
            del line[1]
        if len(line) > 9:
            a = line[1]
            b = line[2]
            c = line[3]
            line[3] = a + ' ' + b + ' ' + c
            del line[1]
            del line[1]
        del line[4]
        del line[7]
        print(','.join(line))
```
